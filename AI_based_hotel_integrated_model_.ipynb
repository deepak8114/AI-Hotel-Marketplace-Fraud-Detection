### Step 2 & 3: Weighting, Scoring, and Normalization for MEI

```python
from sklearn.preprocessing import MinMaxScaler

# Define the features for each pillar
pillar_features = {
    'Platform Trust & Transparency': ['Price_Fairness_Index', 'Guest_Manipulation_Flag'],
    'User Interaction & Engagement': ['Repeat_Booker', 'Cancellation_Ratio', 'Booking_to_Checkin_Days'],
    'Customer Satisfaction': ['Review_Sentiment_Score', 'Review_Length', 'Review_Keyword_Penalty_Flag'],
    'Hotel Performance': ['Hotel_Rating_Avg', 'Verified_Review_Count', 'Fraud_Flag_Count', 'Checkin_Anomaly_Flag'],
    'Pricing & Value': ['Discount_Applied', 'Price_vs_City_Avg', 'Dynamic_Pricing_Flag']
}

# Define the weights for each pillar
pillar_weights = {
    'Platform Trust & Transparency': 0.30,
    'Customer Satisfaction': 0.25,
    'Hotel Performance': 0.20,
    'User Interaction & Engagement': 0.15,
    'Pricing & Value': 0.10
}

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Calculate and normalize sub-scores for each pillar
df['Trust_Transparency_Score'] = 0.0
# For Price_Fairness_Index, values close to 1 are good, >1.1 is penalty.
# We can reverse scale or apply a penalty logic. Let's apply a simple inverse relationship after scaling, or a penalty.
# For simplicity, let's scale it and then handle the penalty.
df['Price_Fairness_Index_Scaled'] = scaler.fit_transform(df[['Price_Fairness_Index']])
# Apply penalty: if index > 1.1, reduce score. This is a simplified penalty.
df['Price_Fairness_Score'] = df['Price_Fairness_Index_Scaled'] * 100
df['Price_Fairness_Score'] = df.apply(lambda row: row['Price_Fairness_Score'] * 0.4 if row['Price_Fairness_Index'] > 1.1 else row['Price_Fairness_Score'], axis=1) # Example penalty

# Guest_Manipulation_Flag: 0 is good, 1 is penalty. Inverse scaling or direct subtraction.
df['Guest_Manipulation_Score'] = (1 - df['Guest_Manipulation_Flag']) * 100 # Inverse scoring

# Combine for Trust & Transparency Score (simple average for now, can be weighted)
df['Trust_Transparency_Score'] = (df['Price_Fairness_Score'] * 0.7 + df['Guest_Manipulation_Score'] * 0.3) # Example internal weighting


df['User_Interaction_Score'] = 0.0
# Repeat_Booker: Assuming 1 is good, 0 is not. Direct scoring.
df['Repeat_Booker_Score'] = df['Repeat_Booker'] * 100

# Cancellation_Ratio: Lower is better. Inverse scaling. Assuming 0 is best.
df['Cancellation_Ratio_Scaled'] = scaler.fit_transform(df[['Cancellation_Ratio']])
df['Cancellation_Score'] = (1 - df['Cancellation_Ratio_Scaled']) * 100

# Booking_to_Checkin_Days: This one is tricky. Maybe closer to a reasonable average is better?
# For simplicity, let's assume more days is slightly better (less last minute), up to a point.
# We can scale and then maybe apply a non-linear transformation or group into bins.
# Let's just scale it directly for now.
df['Booking_to_Checkin_Days_Scaled'] = scaler.fit_transform(df[['Booking_to_Checkin_Days']])
df['Booking_to_Checkin_Days_Score'] = df['Booking_to_Checkin_Days_Scaled'] * 100


# Combine for User Interaction & Engagement Score (simple average for now, can be weighted)
df['User_Interaction_Score'] = (df['Repeat_Booker_Score'] * 0.2 + df['Cancellation_Score'] * 0.4 + df['Booking_to_Checkin_Days_Score'] * 0.4) # Example internal weighting


df['Customer_Satisfaction_Score'] = 0.0
# Review_Sentiment_Score: -1 to +1. Scale to 0-1.
df['Review_Sentiment_Score_Scaled'] = (df['Review_Sentiment_Score'] + 1) / 2 # Scale -1 to 1 range to 0 to 1
df['Review_Sentiment_Score_Normalized'] = scaler.fit_transform(df[['Review_Sentiment_Score_Scaled']]) # Further normalize 0-1 range
df['Review_Sentiment_Score_Final'] = df['Review_Sentiment_Score_Normalized'] * 100

# Review_Length: Longer is generally better, but extremely long could be spam?
# Let's scale it directly, assuming longer is better up to a point, or use bins.
# Direct scaling for simplicity.
df['Review_Length_Scaled'] = scaler.fit_transform(df[['Review_Length']])
df['Review_Length_Score'] = df['Review_Length_Scaled'] * 100

# Review_Keyword_Penalty_Flag: 0 is good, 1 is penalty. Inverse scoring.
df['Review_Keyword_Score'] = (1 - df['Review_Keyword_Penalty_Flag']) * 100

# Combine for Customer Satisfaction Score (simple average for now, can be weighted)
df['Customer_Satisfaction_Score'] = (df['Review_Sentiment_Score_Final'] * 0.6 + df['Review_Length_Score'] * 0.2 + df['Review_Keyword_Score'] * 0.2) # Example internal weighting


df['Hotel_Performance_Score'] = 0.0
# Hotel_Rating_Avg: Direct scaling.
df['Hotel_Rating_Avg_Scaled'] = scaler.fit_transform(df[['Hotel_Rating_Avg']])
df['Hotel_Rating_Score'] = df['Hotel_Rating_Avg_Scaled'] * 100

# Verified_Review_Count: More is better. Direct scaling.
df['Verified_Review_Count_Scaled'] = scaler.fit_transform(df[['Verified_Review_Count']])
df['Verified_Review_Count_Score'] = df['Verified_Review_Count_Scaled'] * 100

# Fraud_Flag_Count: More is worse. Inverse scaling. Assuming 0 is best.
df['Fraud_Flag_Count_Scaled'] = scaler.fit_transform(df[['Fraud_Flag_Count']])
df['Fraud_Flag_Score'] = (1 - df['Fraud_Flag_Count_Scaled']) * 100

# Checkin_Anomaly_Flag: 0 is good, 1 is penalty. Inverse scoring.
df['Checkin_Anomaly_Score'] = (1 - df['Checkin_Anomaly_Flag']) * 100

# Combine for Hotel Performance Score (simple average for now, can be weighted)
df['Hotel_Performance_Score'] = (df['Hotel_Rating_Score'] * 0.4 + df['Verified_Review_Count_Score'] * 0.3 + df['Fraud_Flag_Score'] * 0.15 + df['Checkin_Anomaly_Score'] * 0.15) # Example internal weighting


df['Pricing_Value_Score'] = 0.0
# Discount_Applied: More discount is better. Direct scaling.
# Handle NaN values before scaling, e.g., fill with median or mean, or a value indicating no discount.
df['Discount_Applied'] = df['Discount_Applied'].fillna(0) # Fill NaN with 0 assuming no discount if not specified
df['Discount_Applied_Scaled'] = scaler.fit_transform(df[['Discount_Applied']])
df['Discount_Applied_Score'] = df['Discount_Applied_Scaled'] * 100

# Price_vs_City_Avg: Closer to 1 is better. Can use absolute difference from 1 and inverse scale.
# abs(Price_vs_City_Avg - 1). Lower is better.
df['Price_vs_City_Avg_Difference'] = abs(df['Price_vs_City_Avg'] - 1)
df['Price_vs_City_Avg_Difference_Scaled'] = scaler.fit_transform(df[['Price_vs_City_Avg_Difference']])
df['Price_vs_City_Avg_Score'] = (1 - df['Price_vs_City_Avg_Difference_Scaled']) * 100

# Dynamic_Pricing_Flag: 0 is good, 1 is penalty. Inverse scoring.
df['Dynamic_Pricing_Score'] = (1 - df['Dynamic_Pricing_Flag']) * 100

# Combine for Pricing & Value Score (simple average for now, can be weighted)
df['Pricing_Value_Score'] = (df['Discount_Applied_Score'] * 0.4 + df['Price_vs_City_Avg_Score'] * 0.4 + df['Dynamic_Pricing_Score'] * 0.2) # Example internal weighting


# Calculate the final MEI score using the pillar weights
df['Marketplace_Experience_Index'] = (
    df['Trust_Transparency_Score'] * pillar_weights['Platform Trust & Transparency'] +
    df['Customer_Satisfaction_Score'] * pillar_weights['Customer Satisfaction'] +
    df['Hotel_Performance_Score'] * pillar_weights['Hotel Performance'] +
    df['User_Interaction_Score'] * pillar_weights['User Interaction & Engagement'] +
    df['Pricing_Value_Score'] * pillar_weights['Pricing & Value']
) / 100 # Divide by 100 because pillar scores are 0-100 and weights sum to 1


# Ensure the final MEI is between 0 and 100
df['Marketplace_Experience_Index'] = df['Marketplace_Experience_Index'].clip(0, 100)

# Display the first few rows with the new scores and the final MEI
display(df[['Trust_Transparency_Score', 'User_Interaction_Score', 'Customer_Satisfaction_Score', 'Hotel_Performance_Score', 'Pricing_Value_Score', 'Marketplace_Experience_Index']].head())

# Display summary statistics for the MEI
display(df['Marketplace_Experience_Index'].describe())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# Define features (X) and target (y)
# We'll include engineered features and other relevant columns
feature_cols = [
    'Price', 'Total Rating', 'Rating(out of 5)', 'No of Guest', 'Total Price',
    'Discount (%)', 'Expected Price',
    'Price_Fairness_Index',
    'Repeat_Booker', 'Cancellation_Ratio', 'Booking_to_Checkin_Days',
    'Review_Sentiment_Score', 'Review_Length', 'Review_Keyword_Penalty_Flag',
    'Hotel_Rating_Avg', 'Verified_Review_Count', 'Fraud_Flag_Count', 'Checkin_Anomaly_Flag',
    'Discount_Applied', 'Price_vs_City_Avg',
    'Location', 'Booking Status', 'Payment Mode (online/offline)', 'check in status'
]

# Ensure all feature columns exist in the DataFrame
feature_cols = [col for col in feature_cols if col in df.columns]

X = df[feature_cols]
y = df['Marketplace_Experience_Index']

# Identify categorical and numerical features
categorical_features = X.select_dtypes(include=['object', 'category']).columns
numerical_features = X.select_dtypes(include=np.number).columns

# Create a column transformer for one-hot encoding categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# Apply the preprocessing
X_processed = preprocessor.fit_transform(X)

# Display the shape of the processed features
print("Shape of processed features:", X_processed.shape)
```

### Step 1: Preprocessing - Handle Missing/Null Values

### Step 3: Model Building - Structured Data Anomaly Detection (Autoencoder)

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# Select the structured features for the Autoencoder.
# We'll use the same features as for Isolation Forest and One-Class SVM, excluding the problematic 'Expected Price_Normalized'.
structured_features_for_autoencoder = [
    'Price_Normalized', 'Total Price_Normalized', 'Price_Difference_Percent_Normalized',
    'Review_Sentiment_Score_Normalized', 'Review_Length_Score', 'Review_Keyword_Score',
    'Hotel_Rating_Score', 'Verified_Review_Count_Score',
    'Trust_Transparency_Score', 'User_Interaction_Score', 'Customer_Satisfaction_Score', 'Hotel_Performance_Score', 'Pricing_Value_Score',
    'Hotel ID_Encoded', 'Booking ID_Encoded', 'Customer ID_Encoded'
]

# Ensure all selected structured features exist and handle potential NaNs by imputing
structured_features_for_autoencoder = [col for col in structured_features_for_autoencoder if col in df.columns]
X_structured_autoencoder = df[structured_features_for_autoencoder].copy()

# Impute potential NaNs in the selected structured features if any remain
for col in X_structured_autoencoder.columns:
     if X_structured_autoencoder[col].isnull().any():
         # Use mean imputation as an example
         X_structured_autoencoder[col] = X_structured_autoencoder[col].fillna(X_structured_autoencoder[col].mean())

# Convert to numpy array for Keras
X_structured_autoencoder_np = X_structured_autoencoder.values

# Scale the data for the Autoencoder (important for neural networks)
# We already have some normalized features, but re-scaling the combined set is often beneficial
scaler_autoencoder = MinMaxScaler()
X_scaled_autoencoder = scaler_autoencoder.fit_transform(X_structured_autoencoder_np)

# Split data into training and validation sets for Autoencoder training
# Autoencoders are typically trained on 'normal' data. Since we don't have
# explicitly labeled normal data, we'll train on the majority of our data.
# A common approach is to split the data and assume the training set is mostly normal.
X_train_autoencoder, X_val_autoencoder = train_test_split(
    X_scaled_autoencoder, test_size=0.2, random_state=42
)

print("Shape of training data for Autoencoder:", X_train_autoencoder.shape)
print("Shape of validation data for Autoencoder:", X_val_autoencoder.shape)

# Define the Autoencoder model
input_dim = X_train_autoencoder.shape[1]
encoding_dim = int(input_dim / 2) # Example: encoding dimension is half of input dimension

# Encoder
input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
# Add more encoding layers if needed
# encoder = Dense(int(encoding_dim / 2), activation="relu")(encoder)


# Decoder
decoder = Dense(encoding_dim, activation="relu")(encoder)
# Add more decoding layers if needed
# decoder = Dense(input_dim, activation="relu")(decoder) # Ensure the last layer outputs the original dimension
decoder = Dense(input_dim, activation='sigmoid')(decoder) # Use sigmoid for output scaled between 0 and 1

# Autoencoder model
autoencoder = Model(inputs=input_layer, outputs=decoder)

# Compile the Autoencoder
autoencoder.compile(optimizer='adam', loss='mse') # Mean Squared Error is a common loss for reconstruction

# Train the Autoencoder
# Use the training data (assumed mostly normal) and validate on the validation set
history = autoencoder.fit(X_train_autoencoder, X_train_autoencoder,
                          epochs=50, # Number of training epochs
                          batch_size=32, # Batch size
                          shuffle=True,
                          validation_data=(X_val_autoencoder, X_val_autoencoder),
                          verbose=1) # Set to 0 for less verbose output


# Predict reconstructions for the entire dataset
X_reconstructed = autoencoder.predict(X_scaled_autoencoder)

# Calculate the reconstruction error (e.g., Mean Squared Error)
mse = np.mean(np.power(X_scaled_autoencoder - X_reconstructed, 2), axis=1)

# Use the reconstruction error to identify anomalies
# Define a threshold for anomaly detection. This can be done by:
# 1. Looking at the distribution of reconstruction errors (e.g., using a histogram or box plot)
# 2. Setting a threshold based on a percentile (e.g., top 10% as anomalies)
# 3. Using a statistical method (e.g., based on mean and standard deviation)

# Example: Set threshold based on the mean + 2 * standard deviation
# threshold = np.mean(mse) + 2 * np.std(mse)

# Example: Set threshold based on a percentile (e.g., 95th percentile)
threshold = np.percentile(mse, 95) # Flag the top 5% with highest reconstruction error


df['Autoencoder_Reconstruction_Error'] = mse
df['Autoencoder_Anomaly_Flag'] = np.where(df['Autoencoder_Reconstruction_Error'] > threshold, 1, 0)


print("\nAutoencoder Anomaly Flag Distribution (using 95th percentile threshold):")
display(df['Autoencoder_Anomaly_Flag'].value_counts())

print("\nAutoencoder Reconstruction Errors and Anomaly Flags (higher error is more anomalous):")
display(df[['Autoencoder_Reconstruction_Error', 'Autoencoder_Anomaly_Flag']].head())
```
